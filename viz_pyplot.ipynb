```python
#import potrzebnych bibliotek 
import pandas as pd
import matplotlib.pyplot as plt 

p='./Ludność w Polsce - prognozy.xlsx'
df_ruch=pd.read_excel(p,sheet_name=0) #zaczytanie pierwszego arkusza z pliku
```

```python
df_ruch.head(10) #podgląd danych z pierwszego arkusza 
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

plt.figure(figsize=(22, 6))

# Wykres liniowy dla urodzeń i zgónów
plt.subplot(1, 2, 1)
plt.plot(df_ruch['Rok'], df_ruch['Ruch naturalny – Urodzenia'], label='Urodzenia', marker='o', color='blue')
plt.plot(df_ruch['Rok'], df_ruch['Ruch naturalny – Zgony'], label='Zgony', marker='x', color='red')

plt.twinx()  # Utworzenie drugiej oś y

# Wykres słupkowy dla liczby ludności
plt.bar(df_ruch['Rok'], df_ruch['Ludność na 31 XII'], color='green', alpha=0.5)

# Ustawienie ScalarFormatter dla osi Y drugiego wykresu
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True, useOffset=False))

# Ustawienie etykiet na osiach y
plt.ylabel('Liczba ludności', color='black')
plt.tick_params(axis='y', labelcolor='black')

# Ustawienie tytułu
plt.title('Urodzenia, zgony i liczba ludności w Polsce')

# Dodanie legendy do pierwszego wykresu
plt.legend(loc='upper left')

# Wykres liniowy dla imigracji i emigracji
plt.subplot(1, 2, 2)
plt.plot(df_ruch['Rok'], df_ruch['Migracje zagraniczne na pobyt stały – Imigracja'], label='Imigracja', marker='o', color='blue')
plt.plot(df_ruch['Rok'], df_ruch['Migracje zagraniczne na pobyt stały – Emigracja'], label='Emigracja', marker='x', color='red')

plt.twinx()  # Utworzenie drugiej oś y

# Wykres słupkowy dla salda migracji
plt.bar(df_ruch['Rok'], df_ruch['Migracje zagraniczne na pobyt stały – Saldo'], color='green', alpha=0.5)

# Ustawienie ScalarFormatter dla osi Y drugiego wykresu
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True, useOffset=False))

# Ustawienie etykiet na osiach y
plt.ylabel('Saldo migracji', color='black')
plt.tick_params(axis='y', labelcolor='black')

# Ustawienie tytułu
plt.title('Imigracja, emigracja i saldo migracji')

# Dodanie legendy do drugiego wykresu
plt.legend(loc='upper left')
plt.show()
```

Powyżej udało się stworzyć dwa zaawansowane wykresu, które zawierają w sobie wykresy liniowe dla urodzeń oraz zgonów w pierwszym przypadku oraz wykres słupkowy dla liczby ludności
Analogiczna sytuacja dla imigracji, emigracji oraz salda migracji.

Jednak widzimy, że brakuje tutaj paru zmian kosmetycznych, aby wykres był jeszcze lepszy. 
Brakuje paru szczegółów: 
1) legendy wskazującej, co reprezentuje dana linia na wykresie.
2) aby wykres zaczynał się od początkowego roku z danych
3) aby dane na osi Y w pierwszym wykresie były przedstawione w formie mln 
4) brakuje labeli dla jednej z osi y oraz osi x 
5) można spróbować zmienić kolor na drugim wykresie dla wartości ujemnych na czerwono

```python
plt.figure(figsize=(26, 6)) #zmiana aby poprawić czytelność wykresów

# Wykres liniowy dla urodzeń i zgónów
plt.subplot(1, 2, 1)
plt.plot(df_ruch['Rok'], df_ruch['Ruch naturalny – Urodzenia'], label='Urodzenia', marker='o', color='blue')
plt.plot(df_ruch['Rok'], df_ruch['Ruch naturalny – Zgony'], label='Zgony', marker='x', color='red')
plt.xlabel('Rok')
plt.ylabel("wartość urodzeń/zgonów")
plt.legend(loc='upper center', bbox_to_anchor=(-0.15, 1.0)) #dodanie legendy obok wykresu ze względu na czytelność

plt.twinx()  # Utworzenie drugiej oś y

# Wykres słupkowy dla liczby ludności
plt.bar(df_ruch['Rok'], df_ruch['Ludność na 31 XII'], color='green', alpha=0.5) #dodanie parametru alpha, aby zwiększyć przezroczystość wykresu słupkowego

# Ustawienie ScalarFormatter dla osi Y drugiego wykresu z odpowiednimi parametrami
formatter = ScalarFormatter(useMathText=True, useOffset=False)
formatter.set_scientific(False)  # Wyłączenie notacji naukowej
formatter.set_powerlimits((-3, 3))  # Ograniczenie potęg w notacji naukowej
plt.gca().yaxis.set_major_formatter(formatter)

# Ustawienie etykiet na osiach y
plt.ylabel('Liczba ludności', color='black')
plt.tick_params(axis='y', labelcolor='black')

# Ustawienie etykiet i tytułu
plt.xlabel('Rok', color='black')
plt.title('Urodzenia, zgony i liczba ludności w Polsce - prognoza do 2050')

# Dodanie legendy do pierwszego wykresu
plt.legend(loc='upper left')

# Ustawienie zakresu i kroku na osi x
first_year = df_ruch['Rok'].min()
last_year = df_ruch['Rok'].max()
plt.xticks(range(first_year, last_year + 1, 6))
plt.xlim(first_year - 0.5, last_year + 0.5) # zmiana o 0.5 ponieważ jest to wykres słupkowy 

# Wykres liniowy dla imigracji i emigracji
plt.subplot(1, 2, 2)
plt.plot(df_ruch['Rok'], df_ruch['Migracje zagraniczne na pobyt stały – Imigracja'], label='Imigracja', marker='o', color='blue')
plt.plot(df_ruch['Rok'], df_ruch['Migracje zagraniczne na pobyt stały – Emigracja'], label='Emigracja', marker='x', color='red')
plt.xlabel('Rok')
plt.ylabel("wartość imigracja/emigracja")
plt.legend()

plt.twinx()  # Utworzenie drugiej oś y


saldo = df_ruch['Migracje zagraniczne na pobyt stały – Saldo']
colors = ['red' if val < 0 else 'green' for val in saldo]  # Ustalamy kolory na podstawie wartości

# Wykres słupkowy dla salda migracji
plt.bar(df_ruch['Rok'], df_ruch['Migracje zagraniczne na pobyt stały – Saldo'], color=colors, alpha=0.5)
plt.xlabel('Rok')

# Ustawienie ScalarFormatter dla osi Y drugiego wykresu z odpowiednimi parametrami
formatter = ScalarFormatter(useMathText=True, useOffset=False)

# Ustawienie etykiet na osiach y
plt.ylabel('Saldo migracji', color='black')
plt.tick_params(axis='y', labelcolor='black')

# Ustawienie etykiet i tytułu
plt.xlabel('Rok')
plt.title('Imigracja, emigracja i saldo w Polsce - prognoza do 2050')

# Dodanie legendy do drugiego wykresu
plt.legend(loc='upper left')

# Ustawienie zakresu i kroku na osi x
first_year = df_ruch['Rok'].min()
last_year = df_ruch['Rok'].max()
plt.xticks(range(first_year, last_year + 1, 6))
plt.xlim(first_year - 0.5, last_year + 0.5)
```
